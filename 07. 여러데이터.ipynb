## covid-19 백신 종류별 접종 인구

```python
days= [1,2,3]
az =[2,4,8]
pizer=[5,1,3]
moderna=[1,2,5]

plt.plot(days, az)
plt.plot(days, pizer)
plt.plot(days, moderna)
```

```python
plt.plot(days, az, label = 'az')
plt.plot(days, pizer, label = 'pizer', marker='o', linestyle = '--')
plt.plot(days, moderna, label = 'moderna')
plt.legend(ncol=3)
```
